
### Data Cleaning 
> in this section the unnecssary columns will be deleted 

```python
# delete one unnecssary columns
df.drop('imdb_id', axis=1, inplace=True)
```

```python
# delete one unnecssary columns 
df.drop('id', axis=1, inplace=True)
```

```python
# get the dataset size
df.shape

# show the not null counts of each column
df.info()

# show the data description
display(df.describe())

# find the datatypes of every variable
df.dtypes
```

```python
# run this cell to check everything is in check so far
assert(len(df.columns) == 19)
```

```python
# Handling Missing and Duplicate Values
# list out the nan counts per columns
df.isna().sum()
```

```python
# store column names that has more than 30% data as missing in a list
delete_cols = df.columns[df.count()/df.shape[0] <= .3]
delete_cols
df.drop(delete_cols, axis=1, inplace=True)
df.shape
```

```python
# run this cell to check everything is in check so far
assert(df.shape == (10866, 18))
```

```python
# fill missing values

df.fillna(df.mean(), inplace=True)
df.fillna(df.mode().iloc[0], inplace=True)
df.isna().sum()
```

```python
assert((df.isna().sum() > 0).sum() == 0)
```

```python
# delete duplicate values
df.drop_duplicates(inplace=True)
df.shape
```

```python
# seperate the numeric and categorical columns and store them in different lists

numeric_cols = df.select_dtypes(['int64', 'float64']).columns
numeric_cols

categorical_cols = df.select_dtypes(['object']).columns
categorical_cols
```

```python
# run this cell to check everything is in check so far
assert(len(numeric_cols) == 9)
assert(len(categorical_cols) == 9)
```

```python
# create a dataframe that has only numeric columns
df_numeric = df[numeric_cols]
df_numeric
```

```python
# find all the numeric columns that has less unique values less than 20
numeric_categorical_columns = []
for col in df_numeric.columns:
    if df_numeric[col].nunique() < 20:
        numeric_categorical_columns.append(col)
numeric_categorical_columns
```

### Research Question 2  :WHich geners are most pupular from year to year ?

```python
# typecast numeric_categorical_columns to object

for col in numeric_categorical_columns:
    df[col] = df[col].astype('object')

categorical_cols = list(categorical_cols) + numeric_categorical_columns
```

```python
# run this cell to check everything is in check so far
assert(len(categorical_cols) == 9)
```

```python
# create a new dataframe for analysis
df_categorical = df[categorical_cols]
df_categorical
```

```python
#extract the year from the date 
df['release_date'] = pd.to_datetime(df['release_date'])
df['year'], df['month'] = df['release_date'].dt.year, df['release_date'].dt.month
df
```

```python
#sort the data based on year and popularity
df.groupby(['release_year']).agg({'popularity': 'max'})
```

```python
#drop the duplicate based on the year and keep the last then plot 
temp = df.sort_values(['release_year', 'popularity']).drop_duplicates(['release_year'], keep='last')[['release_year', 'popularity', 'genres']]
temp['xlabel'] = temp.genres + " | " + temp.release_year.astype('str')

plt.rcParams["figure.figsize"] = (20,10)
plt.bar(temp.xlabel, temp.popularity);
plt.xticks(rotation=90);
plt.xlabel('genres across the years');
plt.ylabel('populatiy');
```
